## Question 1: Python/Pandas
Using python and `pandas` methodology and the `township_train.csv` data, create a
table that captures the following:

**A.** The average population for each county: The result should be a `DataFrame` with the columns: `['county_index', 'average_population']`.


```python
df = pd.read_csv('township_train.csv')

#get number of counties
counties = set(df['county_index'])

aves = []
for c in counties:
    pop = []
    #https://stackoverflow.com/questions/16476924/how-to-iterate-over-rows-in-a-dataframe-in-pandas
    #"DataFrame.iterrows is a generator which yields both the index and row (as a Series):"
    for index, row in df.iterrows():
        #print(row['county_index'], row['count_population'])
        if row['county_index'] == c:
            #print(row['count_population'])
            pop.append(row['count_population'])
            
    aves.append(sum(pop)/len(pop))

#Save as dataframe
#https://cmdlinetips.com/2018/01/how-to-create-pandas-dataframe-from-multiple-lists/
DataFrame = pd.DataFrame({'county_index':list(counties),'average_population':aves})

DataFrame
```

**B.** For each municipality type what is the range of breweries per township: The result should be a `DataFrame` with the columns: `['municipality_type', 'min_breweries', 'max_breweries']`

```python
#Repeat process from A
#this time instead of getting average, we save the max and min of each breweries list

df = pd.read_csv('township_train.csv')
munis = set(df['municipality_type'])
maxs = []
mins = []

for m in munis:
    bews = []
    for index, row in df.iterrows():
        if row['municipality_type'] == m:
            bews.append(row['count_breweries'])
            
    maxs.append(max(bews))
    mins.append(min(bews))

DataFrame = pd.DataFrame({'county_index':list(munis),'min_breweries':mins,'max_breweries':maxs})

DataFrame
```

**C.** The number of schools by county: The result should be a `DataFrame` with the columns: `['county_index', 'count_schools']`

```python
#Repeat process from A
#to get school count we have to do some reverse engineering
#and extrapulate from 'count_population' and 'schools_per_10000'

df = pd.read_csv('township_train.csv')
counties = set(df['county_index'])
schools = []

for c in counties:
    s = []
    for index, row in df.iterrows():
        if row['county_index'] == c:
            s.append(int((row['count_population']/10000)*row['schools_per_10000']))
    schools.append(sum(s))
DataFrame = pd.DataFrame({'county_index':list(counties),'count_schools':schools})

DataFrame
```

You are hired as a data scientist by a general contractor that constructs townhome communities. These communities are of various sizes as well as various prices and quality.

# imports

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

## Set up the data

```python
#get Y (our feature of interest in predicting) and X (data about feature)
#https://medium.com/codex/how-to-set-x-and-y-in-pandas-3f38584e9bed
df = pd.read_csv('township_train.csv')
y_col = 'count_townhomes' #feature of interest
y = df[y_col] 
X = df[df.columns.drop(y_col)]


#handle catigorical data
#https://www.kaggle.com/getting-started/27270'
encode_col = 'municipality_type'
X['municipality_type_encoded'] = LabelEncoder().fit_transform(X['municipality_type'])
X = X[X.columns.drop(encode_col)]


# Split features and target into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

#further reading
#https://towardsdatascience.com/data-normalization-with-pandas-and-scikit-learn-7c1cc6ed6475
```

**A.** Construct a model that is capable of predicting the number of townhomes in each township.

Include any Exploratory Data Analysis (EDA), feature engineering and model training code here. Feel free to organize the contents


```python
RF = RandomForestRegressor()
RF.fit(X_train, y_train);
```

**B.** Provide evaluations of the model's performance

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html
#https://stackoverflow.com/questions/50789508/random-forest-regression-how-do-i-analyse-its-performance-python-sklearn
#https://cnvrg.io/random-forest-regression/

pred = RF.predict(X_test)
print('MAE: ', mean_absolute_error(y_test, pred))
print('MSE: ', mean_squared_error(y_test, pred)) 
```

**A&B alternative**

Same expirment with support vector machine doesn't do as well

```python
#Support vector machine
from sklearn.svm import SVR
#https://medium.com/pursuitnotes/support-vector-regression-in-6-steps-with-python-c4569acd062d
#also has good info on feature cleaning
regressor = SVR(kernel='rbf')
regressor.fit(X,y)
pred = regressor.predict(X_test)
print('MAE: ', mean_absolute_error(y_test, pred))
print('MSE: ', mean_squared_error(y_test, pred))
```


# Enhanced Recommendations for Business Insights

After rigorous analysis and modeling, the insights gleaned offer valuable considerations for the strategic planning and development of new townhome communities. Here, we distill those findings into actionable recommendations:

## A. Location Selection for New Townhome Communities

- **Market Demand and Supply Analysis:** The number of existing townhomes in a county serves as a proxy for both market demand and supply saturation. High concentrations suggest robust demand but may also indicate market saturation. Conversely, areas with fewer townhomes than predicted could represent untapped markets. It is crucial to balance between market demand and avoiding oversaturation.

- **Demographic and Economic Factors:** Consider the average population, median income, and municipal budget of counties. Higher incomes and budgets may support premium townhome developments, while areas with growing populations could indicate a rising demand for housing.

- **Amenities and Infrastructure:** The presence of parks, schools, and breweries can attract specific demographics. For instance, areas with numerous parks and schools may appeal more to families, whereas a high count of breweries might attract younger adults or couples.

## B. Recommendations on Townhome Quality and Pricing

- **Matching Community Preferences:** The quality and pricing of new townhomes should align with the municipality type's characteristics and resident preferences. Urban areas might demand higher-quality, premium-priced townhomes due to higher median incomes and lifestyle expectations. In contrast, rural areas might prioritize affordability and value for money.

- **Competitive Analysis:** Investigate the quality and pricing of existing townhomes within the target area. Strive to offer competitive advantages, whether through superior quality, amenities, or pricing strategies, to attract potential buyers.

- **Future Value Proposition:** Consider future developments in the area that could affect property values. Areas slated for economic development or infrastructure upgrades can be prime locations for higher-quality townhomes, anticipating future value appreciation.

## C. Additional Data for Enhanced Model Precision

- **Detailed Demographics:** Information on family sizes, age distribution, and employment sectors could refine predictions on the type and size of townhomes in demand.

- **School Quality and Business Profiles:** Data on the quality of education and the mix of local businesses could provide insights into the community's priorities and lifestyle preferences, aiding in targeted development.

- **Geospatial Analysis:** Utilizing GIS data could reveal insights on land use patterns, proximity to key amenities, and environmental factors that significantly impact location desirability.

- **Consumer Preferences Survey:** Direct feedback on preferences and expectations from potential buyers in the target demographics could guide the development of tailored townhome communities.

- **Market Trends Analysis:** Current and projected real estate market trends should inform the timing, scale, and nature of development projects to align with market cycles and investor expectations.

**Conclusion:** The integration of these considerations and data into the strategic planning process can enhance the precision of market analyses, ensure developments meet market needs, and maximize the investment return for new townhome communities.
